**NumPy (Numerical Python)** é a base para computação numérica em Python. Ele fornece um tipo de array multidimensional (numpy.ndarray) mais rápido que listas do Python, além de funções matemáticas, álgebra linear, transformadas fourier, etc..

**Pandas** é uma extensão do Numpy que facilita o trabalho com dados tabulares (planilhas, CSVs, banco de dados). Ele oferece as estruturas Series (uma coluna) e DataFrame (tabela) com poderosas operaçõies de filtragem, agregação, junção e visualização de dados.

**Matplotlib** é a biblioteca de plotagem mais comum em Python. Ela cria gráficos 2-D (linhas, barras, dispersão, histogra,a, etc..) a partir de dados em arrays ou DataFrames.

Essas três bibliotecas formam o core do ecossistema de ciência de dados e análise estatística em Python. Aprender a usá-las permite que você:

1. Manipule grandes volumes de dados de forma eficiente (NumPy) e com estrutura tabular (Pnadas)
2. Realize análises exploratórios (estatísticas, agregações) e visualize resultados (Matplotlib)
3. Construa pipelines de dados que podem ser usados em machine learning, relatórios e dashboards

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np_data = np.random.normal(loc=0, scale=1, size=1000)
```

```python
df = pd.DataFrame({'VALOR': np_data})
```

```python
media = df['VALOR'].mean()
desvio = df['VALOR'].std()

print(f"media = {media:.3f} | desvio padrão = {desvio:.3f}")
```

```python
plt.hist(df['VALOR'], bins=30, color='steelblue', edgecolor='black')
plt.title('Distribuição normal (1000 amostras)')
plt.xlabel('VALOR')
plt.ylabel('Frequência')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show
```

**Explicação de cada código**

Importações

numpy é importado como np por convenção.
pandas como pd.
matplotlib.pyplot como plt (interface de plotagem estilo MATLAB).
Criação de dados

np.random.normal gera 1 000 números aleatórios distribuídos normalmente (média 0, desvio 1).
O resultado é um numpy.ndarray.
DataFrame

pd.DataFrame({'valor': np_data}) transforma o array em uma tabela com uma coluna chamada valor.
DataFrame permite operações como mean() e std() diretamente.
Estatísticas

df['valor'].mean() calcula a média dos valores.
df['valor'].std() calcula o desvio padrão.
Plotagem

plt.hist cria um histograma com 30 bins.
color, edgecolor, title, xlabel, ylabel definem a aparência.
plt.grid adiciona linhas de grade para facilitar a leitura.
plt.show() exibe o gráfico.
